```python
url='https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'

fd_data = pd.read_csv(url, encoding= 'latin_1')
fd_data.describe(include = 'all')
```

```python
#Using the mean of each column to fill the NaN of that column
y = fd_data[['Y2014','Y2015','Y2016','Y2017','Y2018']].mean()
print(y)

fd_data[['Y2014','Y2015','Y2016','Y2017','Y2018']]=fd_data[['Y2014','Y2015','Y2016',
         'Y2017','Y2018']].fillna(value= y)
```
